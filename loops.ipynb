Break
To stop before integer

```python
b=[1,2,3,90,'divya','arunachalam', 7+8j, True,False]
for i in b:
    if type(i)==str:
        break
    print(i)
```

Continue

when specific condition met, it allows to skip some part of statement and control transfer to next statement

```python
b=[1,2,3,90,'divya','arunachalam', 7+8j, True,False]
for i in b:
    if type(i)==str:
        continue
    print(i)
```

```python
b=[1,2,3,90,'divya','arunachalam', 7+8j, True,False]
for i in b:
    if type(i)==int:
        continue
    print(i)
```
